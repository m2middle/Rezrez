```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import joblib
```

```python
# Load test data features csv
test_data_features = pd.read_csv("student_data/test_data_features.csv")
test_data_features.head()
```

```python
# load the model and encoders
model, encoders = joblib.load("student_model.pkl")

# Encode the categoricals variable using the encoder object
categorical_cols = ["under_18", "famsize", "schoolsup", "selfpayfees", "activities", "catch-up", "internet", "romantic", "specialty", "sex", "Mjob", "Fjob", "reason", "socialcohabitation"]

for i, encoder in enumerate(encoders):
    col = categorical_cols[i]
    test_data_features[col] = encoder.transform(test_data_features[col])

test_data_features.head()
```

```python
predictions = model.predict(test_data_features.values)
```

```python
list_students = []
for i in range(len(predictions)):
    list_students.append("Student1" + str(i+1))
```

```python
output_df = pd.DataFrame({"Student": list_students, "Prediction": predictions})
output_df["Status"] = np.where(output_df["Prediction"] == 1, "pass", "fail")
output_df.head()
```
