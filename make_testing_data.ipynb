```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('student_data/student_mat_new.csv', sep=',')
```

```python
# Make a test data which is 10% of the orignial dataframe
test_data = data.sample(frac=0.1, random_state=42)
```

```python
# Save test data to csv
test_data.to_csv('student_data/test_data.csv', index=False)  
```

```python
test_data_new = pd.read_csv("student_data/test_data.csv")
test_data_new.head() 
```

```python
test_data_new["G3"] = np.where(test_data_new["G3"] >= 10, 1, 0)
```

```python
# Save the values in "G3" to a txt file line by line
test_data_new["G3"].to_csv("student_data/test_data_real_passfail.txt", index=False, header=False)
```

```python
# Save all columns except "G3" into a new csv named "test_data_features.csv"  
test_data_new.drop("G3", axis=1).to_csv("student_data/test_data_features.csv", index=False)
```
